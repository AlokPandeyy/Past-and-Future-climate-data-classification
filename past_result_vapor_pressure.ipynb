```python
import pandas as pd

# Constants
P = 101325  # Pressure in Pascals (1 atm)

# Function to calculate water vapor pressure
def calculate_vapor_pressure(T, RH):
    """
    Calculate water vapor pressure using the Magnus formula.
    T: Dry bulb temperature in °C
    RH: Relative humidity (0 to 1)
    """
    # Magnus formula constants for water vapor pressure over liquid water
    A = 611.21  # Pa
    B = 17.502
    C = 240.97  # °C
    
    # Calculate saturation vapor pressure
    saturation_vapor_pressure = A * (2.71828 ** ((B * T) / (C + T)))
    
    # Calculate actual vapor pressure
    vapor_pressure = saturation_vapor_pressure * RH
    
    return vapor_pressure

# Ensure your columns are named correctly
# Replace 'mean_dbt' and 'mean_rh' with your actual column names
df['(January vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(January, mean_dbt)'], row['(January, mean_rh)']/100), axis=1)
df['(February vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(February, mean_dbt)'], row['(February, mean_rh)']/100), axis=1)
df['(March vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(March, mean_dbt)'], row['(March, mean_rh)']/100), axis=1)
df['(April vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(April, mean_dbt)'], row['(April, mean_rh)']/100), axis=1)
df['(May vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(May, mean_dbt)'], row['(May, mean_rh)']/100), axis=1)
df['(June vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(June, mean_dbt)'], row['(June, mean_rh)']/100), axis=1)
df['(July vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(July, mean_dbt)'], row['(July, mean_rh)']/100), axis=1)
df['(August vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(August, mean_dbt)'], row['(August, mean_rh)']/100), axis=1)
df['(September vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(September, mean_dbt)'], row['(September, mean_rh)']/100), axis=1)
df['(October vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(October, mean_dbt)'], row['(October, mean_rh)']/100), axis=1)
df['(November vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(November, mean_dbt)'], row['(November, mean_rh)']/100), axis=1)
df['(December vapor_pressure)'] = df.apply(lambda row: calculate_vapor_pressure(row['(December, mean_dbt)'], row['(December, mean_rh)']/100), axis=1)
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=5)
X_pca = pca.fit_transform(data_scaled)
pca.explained_variance_ratio_
np.cumsum(pca.explained_variance_ratio_)
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
import seaborn as sns

# Load PCA-transformed data
# X_pca = np.load("pca_transformed_data.npy")

# Define range for number of clusters and fuzziness parameter
c_values = np.arange(2, 36, 1)  # c from 2 to 35
m_values = np.arange(1.4, 2.3, 0.1)  # m from 1.4 to 2.2 (increment 0.1)

# Initialize matrix to store CVIXB values (rows: m, cols: c)
cvixb_matrix = np.zeros((len(m_values), len(c_values)))

# Compute CVIXB for each (c, m)
for i, m in enumerate(m_values):
    for j, c in enumerate(c_values):
        cntr, u, _, _, _, _, _ = fuzz.cluster.cmeans(
            X_pca.T, c, m, error=1e-6, maxiter=1000, init=None
        )

        # Compute distances between data points and cluster centers
        distances = np.linalg.norm(X_pca[:, None, :] - cntr[None, :, :], axis=2)

        # Compute within-cluster scatter
        scatter = np.sum((u.T ** 2) * (distances ** 2))

        # Compute minimum inter-cluster distance
        pairwise_distances = np.linalg.norm(cntr[:, None, :] - cntr[None, :, :], axis=2)
        np.fill_diagonal(pairwise_distances, np.inf)  # Ignore self-distance
        min_dist = np.min(pairwise_distances)

        # Compute CVIXB
        cvixb_matrix[i, j] = scatter / (X_pca.shape[0] * min_dist ** 2)

# Find optimal (c, m) with lowest CVIXB
optimal_idx = np.unravel_index(np.argmin(cvixb_matrix), cvixb_matrix.shape)
optimal_c = c_values[optimal_idx[1]]
optimal_m = m_values[optimal_idx[0]]

# Get **top 50** optimal values (smallest CVIXB scores)
flat_indices = np.argsort(cvixb_matrix.flatten())[:50]  # Select top 50 lowest CVIXB
top_50_indices = np.unravel_index(flat_indices, cvixb_matrix.shape)
top_50_c_values = c_values[top_50_indices[1]]
top_50_m_values = m_values[top_50_indices[0]]
top_50_scores = cvixb_matrix[top_50_indices]

# Print optimal results
print(f"Optimal Number of Clusters (c): {optimal_c}")
print(f"Optimal Fuzziness (m): {optimal_m}")

print("\nTop 50 Optimal (c, m) values based on CVIXB:")
for rank, (c_opt, m_opt, score) in enumerate(zip(top_50_c_values, top_50_m_values, top_50_scores), 1):
    print(f"Rank {rank}: c = {c_opt}, m = {m_opt:.1f}, CVIXB Score = {score:.4f}")
```

```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz

# Load the original dataset
df_original = pd.read_excel("vapor_pressure_past_climate_data_4variables.xlsx")

# Perform Fuzzy C-Means Clustering
c = 14   # Number of clusters
m = 1.6  # Fuzzifier
error = 1e-6
maxiter = 1000

cntr, u, _, _, _, _, _ = fuzz.cluster.cmeans(
    X_pca.T, c, m, error=error, maxiter=maxiter, init=None
)

# Assign cluster labels based on the highest membership probability
cluster_labels = np.argmax(u, axis=0)  # Shape: (n_samples,)

# Compute confidence scores (max probability per row)
confidence_scores = np.max(u, axis=0)

# Convert membership matrix to DataFrame
membership_df = pd.DataFrame(u.T, columns=[f"Cluster_{i}_Probability" for i in range(c)])

# Fix: Convert DataFrame to NumPy array before sorting
sorted_probs = np.sort(membership_df.values, axis=1)  # Sort probabilities per row
second_highest_probs = sorted_probs[:, -2]  # Second highest probability

# Compute alternative confidence score
confidence_scores_1 = 1 - second_highest_probs

# Add cluster labels and confidence scores to the original DataFrame
df_original["Cluster"] = cluster_labels
df_original["Confidence_Score"] = confidence_scores
df_original["Confidence_Score_1"] = confidence_scores_1  # Ensure this is added correctly

# Merge membership probabilities with original data
df_final = pd.concat([df_original, membership_df], axis=1)

# Save the final dataset with cluster labels, confidence scores, and probabilities
df_final.to_excel("fuzzy_clustered_data.xlsx", index=False)

# Print sample output
print(df_final[["Cluster", "Confidence_Score", "Confidence_Score_1"]].head())
```

```python
import pandas as pd

# Print column names to verify
print(df_original.columns)

# Define month names
months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

# Create column lists based on actual names
dbt_min_cols = [f"({month}, min_dbt)" for month in months]
dbt_max_cols = [f"({month}, max_dbt)" for month in months]
vp_cols = [f"({month} vapor_pressure)" for month in months]
ghi_cols = [f"({month}, mean_ghi)" for month in months]

# Define aggregation functions
aggregation_functions = {"Cluster": "count"}  # Number of rows per cluster

# Function to get min, max, and mean across multiple columns
def min_across_months(x):
    return x.min(axis=1).min()

def max_across_months(x):
    return x.max(axis=1).max()

def mean_across_months(x):
    return x.mean(axis=1).mean()

# Add aggregate calculations for each variable type
for cols, var_name in zip([dbt_min_cols, dbt_max_cols, vp_cols, ghi_cols], 
                        ["Min_DBT", "Max_DBT", "Vapor_Pressure", "GHI"]):
    aggregation_functions[f"Min_{var_name}"] = (lambda df: min_across_months(df[cols]))
    aggregation_functions[f"Max_{var_name}"] = (lambda df: max_across_months(df[cols]))
    aggregation_functions[f"Avg_{var_name}"] = (lambda df: mean_across_months(df[cols]))

# Apply aggregation functions to grouped DataFrame
cluster_stats = df_original.groupby("Cluster").apply(lambda df: pd.Series({
    "Row Count": len(df),
    "Min Min_DBT": min_across_months(df[dbt_min_cols]),
    "Max Min_DBT": max_across_months(df[dbt_min_cols]),
    "Min Max_DBT": min_across_months(df[dbt_max_cols]),
    "Max Max_DBT": max_across_months(df[dbt_max_cols]),
    "Avg Min_DBT": mean_across_months(df[dbt_min_cols]),
    "Avg Max_DBT": mean_across_months(df[dbt_max_cols]),
    "Min Vapor_Pressure": min_across_months(df[vp_cols]),
    "Max Vapor_Pressure": max_across_months(df[vp_cols]),
    "Avg Vapor_Pressure": mean_across_months(df[vp_cols]),
    "Min GHI": min_across_months(df[ghi_cols]),
    "Max GHI": max_across_months(df[ghi_cols]),
    "Avg GHI": mean_across_months(df[ghi_cols])
}))

# Save to Excel
cluster_stats.to_excel("cluster_statistics.xlsx")

# Print result
print(cluster_stats)
```
