```python
import matplotlib.pyplot as plt
import numpy as np

# Number of PCs and features
num_pcs = factor_loadings.shape[0]
num_features = factor_loadings.shape[1]

# Function to create radar plot
def radar_plot(ax, factor_loadings, feature_names):
    # Number of variables (features)
    num_vars = len(feature_names)

    # Compute the angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # Make the radar plot circular by repeating the first value for each PC
    factor_loadings = np.hstack((factor_loadings, factor_loadings[:, [0]]))  # Add the first factor loading at the end
    angles += angles[:1]

    # Plot each principal component without filling
    for pc_num in range(num_pcs):
        ax.plot(angles, factor_loadings[pc_num, :], label=f'PC{pc_num + 1}', linewidth=2)

    # Add feature labels at the correct angles
    for i in range(num_vars):
        ax.text(angles[i], factor_loadings[0, i] * 1.1, feature_names[i], color='red', ha='center', va='center', fontsize=10)

    # Set the title and labels
    ax.set_title('Radar Plot for All 5 PCs', size=14)
    ax.set_ylim(-1.2, 1.2)  # Adjust the limits if necessary
    ax.set_yticklabels([])  # Hide radial ticks
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(feature_names, size=12)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Create the radar plot for all PCs
radar_plot(ax, factor_loadings, data.columns)

# Display the legend
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skfuzzy import cmeans
from sklearn.metrics import silhouette_score

# Load PCA transformed data (already scaled and reduced to 5 components)
data_pca = np.load("pca_transformed_data.npy")  # Replace with actual PCA output

cluster_range = range(10, 20)
m_values = np.arange(1.5, 2.6, 0.1)
max_iter_range = range(1000, 31000, 5000)

results = []
best_score = -1
best_params = None

for n_clusters in cluster_range:
    for m in m_values:
        for max_iter in max_iter_range:
            cntr, u, u0, d, jm, p, fpc = cmeans(
                data_pca.T, c=n_clusters, m=m, error=0.005, maxiter=max_iter, init=None
            )
            labels = np.argmax(u, axis=0)
            silhouette_avg = silhouette_score(data_pca, labels)
            results.append((n_clusters, m, max_iter, silhouette_avg))

            if silhouette_avg > best_score:
                best_score = silhouette_avg
                best_params = (n_clusters, m, max_iter)

results_df = pd.DataFrame(results, columns=['n_clusters', 'm', 'max_iter', 'silhouette_score'])

# Print the optimal parameters
print(f"Best Parameters: Clusters={best_params[0]}, m={best_params[1]}, max_iter={best_params[2]}")
print(f"Best Silhouette Score: {best_score}")

# Plotting
grouped = results_df.groupby('n_clusters')
for name, group in grouped:
    plt.plot(group['m'], group['silhouette_score'], marker='o', linestyle='-', label=f'Clusters: {name}')

plt.xlabel('Fuzziness parameter (m)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs Fuzziness Parameter for Different Cluster Numbers')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skfuzzy import cmeans

# Load PCA transformed data (already scaled and reduced to 5 components)
data_pca = np.load("pca_transformed_data.npy")  # Replace with actual PCA output

# Define grid search parameters
cluster_range = range(6, 10)  # Number of clusters
m_values = np.arange(1.5, 2.6, 0.1)  # Fuzziness parameter
max_iter_range = range(1000, 31000, 5000)  # Max iterations

results = []
best_fpc = -1
best_params = None

# Perform grid search
for n_clusters in cluster_range:
    for m in m_values:
        for max_iter in max_iter_range:
            cntr, u, _, _, _, _, fpc = cmeans(
                data_pca.T, c=n_clusters, m=m, error=0.005, maxiter=max_iter, init=None
            )

            results.append((n_clusters, m, max_iter, fpc))

            # Track best parameters based on highest FPC
            if fpc > best_fpc:
                best_fpc = fpc
                best_params = (n_clusters, m, max_iter)

# Convert results to a DataFrame
results_df = pd.DataFrame(results, columns=['n_clusters', 'm', 'max_iter', 'fpc'])

# Print the optimal parameters
print(f"Best Parameters: Clusters={best_params[0]}, m={best_params[1]}, max_iter={best_params[2]}")
print(f"Best FPC Score: {best_fpc}")

# Visualization
grouped = results_df.groupby('n_clusters')
for name, group in grouped:
    plt.plot(group['m'], group['fpc'], marker='o', linestyle='-', label=f'Clusters: {name}')

plt.xlabel('Fuzziness parameter (m)')
plt.ylabel('Fuzzy Partition Coefficient (FPC)')
plt.title('FPC vs Fuzziness Parameter for Different Cluster Numbers')
plt.legend()
plt.show()
```

```python
from sklearn.base import BaseEstimator, ClassifierMixin
import numpy as np
from fcmeans import FCM

class FuzzyCMeansClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, n_clusters=3, max_iter=150, m=2.0, error=1e-5):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.m = m
        self.error = error

    def fit(self, X, y=None):
        # Fit the FCM model
        self.fcm = FCM(n_clusters=self.n_clusters, max_iter=self.max_iter, m=self.m)
        self.fcm.fit(X)
        
        # The cluster centers and membership matrix
        self.cntr = self.fcm.centers
        self.u = self.fcm.u
        
        # Assign labels based on highest membership
        self.labels_ = np.argmax(self.u, axis=0)
        return self

    def predict(self, X):
        # Predict the membership matrix for the new data
        u = self.fcm.predict(X)
        
        # Assign labels based on highest membership
        return np.argmax(u, axis=1)
```
