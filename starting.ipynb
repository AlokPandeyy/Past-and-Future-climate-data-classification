```python
# Group the data by the "Month" column
monthly_statistics = {}

for month, group in columns.groupby("Month"):
    stats = {}
    
    # Calculate the mean, min, max, and 99th percentiles for each column
    if "Dry Bulb Temperature [C]" in group:
        stats["mean_dbt"] = group["Dry Bulb Temperature [C]"].mean()
        stats["min_dbt"] = group["Dry Bulb Temperature [C]"].min()
        stats["max_dbt"] = group["Dry Bulb Temperature [C]"].max()
    
    if "Relative Humidity [%]" in group:
        stats["mean_rh"] = group["Relative Humidity [%]"].mean()
    
    if "Global Horizontal Radiation [Wh/m2]" in group:
        stats["mean_ghi"] = group[group["Global Horizontal Radiation [Wh/m2]"] > 50]["Global Horizontal Radiation [Wh/m2]"].mean()
        stats["99th_percentile_ghi"] = group["Global Horizontal Radiation [Wh/m2]"].quantile(0.99)
    
    if "Direct Normal Radiation [Wh/m2]" in group:
        stats["mean_drsr"] = group[group["Direct Normal Radiation [Wh/m2]"] > 50]["Direct Normal Radiation [Wh/m2]"].mean()
        stats["99th_percentile_drsr"] = group["Direct Normal Radiation [Wh/m2]"].quantile(0.99)
    
    if "Diffuse Horizontal Radiation [Wh/m2]" in group:
        stats["mean_dfhr"] = group[group["Diffuse Horizontal Radiation [Wh/m2]"] > 50]["Diffuse Horizontal Radiation [Wh/m2]"].mean()
        stats["99th_percentile_dfhr"] = group["Diffuse Horizontal Radiation [Wh/m2]"].quantile(0.99)
    
    if "Wind Speed [m/s]" in group:
        stats["mean_ws"] = group["Wind Speed [m/s]"].mean()
    
    # Store the stats for the current month
    monthly_statistics[f"Month_{month}"] = stats

# Print the monthly statistics
for month, stats in monthly_statistics.items():
    print(f"Statistics for {month}:")
    print(stats)
```

```python
month_names = [
    "January", "February", "March", "April", "May", "June", 
    "July", "August", "September", "October", "November", "December"
]

# Flatten the monthly_statistics dictionary into a single row
columns = ["year"]  # The first column for "year"
values = [""]  # Leaving the first cell blank as per your request

for month, stats in monthly_statistics.items():
    # Extract the numeric part of the month (e.g., "Month_1" becomes 1)
    month_number = int(month.split("_")[1])  # This will give the month number (1, 2, ..., 12)
    month_name = month_names[month_number - 1]  # Get the month name
    for stat_name, value in stats.items():
        # Create column names in the format "(Month, stat_name)"
        columns.append(f"({month_name}, {stat_name})")
        values.append(value)

# Save to Excel
df = pd.DataFrame([values], columns=columns)
df.to_excel("monthly_statistics.xlsx", index=False)
```
